```python
ad_clicks = pd.read_csv('ad_clicks.csv')
ad_clicks.head()
```

```python
utm_source = ad_clicks.groupby('utm_source')\
         .user_id.count().reset_index()
utm_source
```

```python
ad_clicks['is_click'] = ~ad_clicks\
   .ad_click_timestamp.isnull()
ad_clicks.head()
```

```python
clicks_by_source = ad_clicks\
   .groupby(['utm_source',
             'is_click'])\
   .user_id.count()\
   .reset_index()
clicks_by_source
```

```python
clicks_pivot = clicks_by_source.pivot(columns='is_click',index='utm_source',
 values='user_id').reset_index()
```

```python
clicks_pivot['percent_clicked'] = (clicks_pivot[True]/(clicks_pivot[True]+\
                     clicks_pivot[False]))\
* 100
```

```python
clicks_pivot
```
